```python
import json
import csv
import pandas

errors = 0
data = []
with open('../data/twitter/sample.ndjson') as fp:
    for line in fp.readlines():
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError:
            errors += 1

print(errors)
print(len(data))
```

```python
print(json.dumps(data[0], indent=2))
```

```python
with open('../data/twitter/hc_sample.csv', 'w+') as fp:
    fieldnames = ['id_str', 'reply_to', 'text']
    writer = csv.DictWriter(fp, fieldnames=fieldnames)
    writer.writeheader()
    for d in data:
        if 'extended_tweet' in d:
            text = d['extended_tweet']['full_text']
        else:
            text = d['text']
        text = repr(text)[1:-1]
        writer.writerow(dict(id_str=d['id_str'], 
                             text=repr(text), 
                             reply_to=d['in_reply_to_screen_name']))
```
